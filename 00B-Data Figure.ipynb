```python
import pickle
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"

import string
```

```python
POLLUTANT = "PM2.5"
```

```python
cities = ['delhi', 'seoul', 'skopje', 'ulaanbaatar']
```

```python
fig, axs = plt.subplots(nrows=len(cities), figsize=(8, 3*len(cities)))
for idx, city in enumerate(cities):
    
    dataset = pickle.load(open(f"./Data/{cities[idx]}_dataset.pkl", "rb"))
    station = dataset[0]
    station['df'].plot(ax=axs[idx], c='k')
    axs[idx].set_ylabel(f"{city.capitalize()}\nPM2.5", fontsize=16)
    axs[idx].set_xlabel('', fontsize=18)
    axs[idx].grid(ls='--')
    axs[idx].tick_params(axis='both', which='major', labelsize=16)
    axs[idx].tick_params(axis='both', which='minor', labelsize=16)
    axs[idx].get_legend().remove()
    
    axs[idx].text(0.45, -0.45, f"({string.ascii_lowercase[idx]})", transform=axs[idx].transAxes, 
            size=20, weight='bold')
    
plt.subplots_adjust(hspace=0)
# plt.suptitle('Time-Series Data', y=1.02, fontsize=20)
plt.tight_layout()
plt.savefig(f"./paper_figures/ts-data.png", dpi=400, bbox_inches='tight')
```
