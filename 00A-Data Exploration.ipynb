```python
import pickle
import pandas as pd

import matplotlib.pyplot as plt
```

```python
POLLUTANT = "PT08.S5(O3)"
```

```python
dataset = pickle.load(open("./Data/air_quality_uci_dataset.pkl", "rb"))
```

```python
dataset.columns
```

```python
def plot_station(df, station_name):
    fig = plt.figure(figsize=(20, 4))
    ax = fig.gca()
    
    df.plot(marker='o', color='k', markersize=1, ax=ax)
    plt.grid(ls='--')
    ax.set_axisbelow(True)
    plt.ylabel(f"{POLLUTANT}")
    plt.xlabel("Date")

    plt.title(f"{station_name}")
```

```python
plot_station(dataset[POLLUTANT], 'Air Quality UCI')
```
